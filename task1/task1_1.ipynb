```python
import numpy as np
import matplotlib.pyplot as plt
low = 0
high = 1
N = 1000
x = np.random.uniform(low, high, N)
plt.plot(x, '.', markersize=3)
plt.xlabel("Sample index")
plt.ylabel("Value")
plt.title("Uniformly Distributed Samples")
plt.show()
```

```python
theoretical_mean = (low + high)/2
sample_mean = x.sum()/N
print("Theoretical Mean: ", theoretical_mean)
print("Sample Mean: ", sample_mean)
print("Mean error: ", theoretical_mean - sample_mean)
```

Sample mean is a good estimator of the theoretical mean, as all samples come from the same underlying time invariant distribution.

```python
# theoretical_SD = (high-low)/np.sqrt(12)
theoretical_SD = np.sqrt(np.mean(((x - 0.5) ** 2)))
sample_SD = np.sqrt(1/(N-1) * (np.sum(np.square(x-sample_mean))))
print("Theoretical SD: ", theoretical_SD)
print("Sample SD: ", sample_SD)
print("SD error: ", sample_SD - theoretical_SD)
```

Sample estimate of standard deviation is also accurate due to the high number of samples used. 

```python
M = 10
N = 1000
x = np.random.uniform(low, high, size=(N, M))
theoretical_mean = (low + high)/2
sample_mean = np.sum(x, axis=0)/N
# theoretical_SD = (high-low)/np.sqrt(12)
theoretical_SD = np.sqrt(np.mean(((x - 0.5) ** 2)))
sample_SD = np.sqrt(1/(N-1) * (np.sum(np.square(x-sample_mean), axis=0)))
plt.plot(sample_mean, '.', markersize=3)
plt.axhline(theoretical_mean, color='red', linestyle='--', linewidth=1)
plt.xlabel("Realisation Index")
plt.ylabel("Sample mean")
plt.title("Sample means over 10 realisations")
plt.show()

plt.plot(sample_SD, '.', markersize=3)
plt.axhline(theoretical_SD, color='red', linestyle='--', linewidth=1)
plt.xlabel("Realisation Index")
plt.ylabel("Sample standard deviation")
plt.title("Sample standard deviation over 10 realisations")
plt.show()

```

Comment about their bias here. 

```python
x_hist = np.random.uniform(low, high, N*100)
theoretical_pdf = 1
plt.hist(x_hist, bins=15, density=True, color='blue', edgecolor='black')
plt.axhline(theoretical_pdf, color='red', linestyle='--', linewidth=1)
plt.xlabel('Value')
plt.ylabel('pdf')
plt.title('Theoretical and Estimate pdf')
plt.show()
```

The estimate appears to converge as the number of generated samples increases (law of large numbers). As number of bins increases, we get a less smooth estimate pdf with more variance.

```python
import random as rng
M = 10
N = 1000
rng = np.random.default_rng()
x = rng.standard_normal((N, M))
theoretical_mean = 0
sample_mean = np.sum(x, axis=0)/N
theoretical_SD = 1
sample_SD = np.sqrt(1/(N-1) * (np.sum(np.square(x-sample_mean), axis=0)))
plt.plot(sample_mean, '.', markersize=3)
plt.axhline(theoretical_mean, color='red', linestyle='--', linewidth=1)
plt.xlabel("Realisation Index")
plt.ylabel("Sample mean")
plt.title("Sample means over 10 realisations")
plt.show()

plt.plot(sample_SD, '.', markersize=3)
plt.axhline(theoretical_SD, color='red', linestyle='--', linewidth=1)
plt.xlabel("Realisation Index")
plt.ylabel("Sample standard deviation")
plt.title("Sample standard deviation over 10 realisations")
plt.show()
```
